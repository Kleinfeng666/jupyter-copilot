```julia
using Plots

# Define the Lissajous curve
function lissajous_curve(a, b, δ, t)
    x = sin.(a .* t .+ δ)
    y = sin.(b .* t)
    return x, y
end

function

# Generate data points for the curve
t = range(0, stop=2π, length=1000)
a = 5
b = 4
δ = π / 2

x, y = lissajous_curve(a, b, δ, t)

# Plot the Lissajous curve
plot(x, y, linecolor=:blue, lw=2, title="Lissajous Curve", xlabel="X", ylabel="Y")
```
