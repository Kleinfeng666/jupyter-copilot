```julia
using Plots
using LinearAlgebra

function julia_set(z, c, max_iter)
    for n = 1:max_iter
        if abs(z) > 2
            return n
        end
        z = z^2 + c
    end
    return max_iter
end


function create_julia_fractal(width, height, c, max_iter)
    x = range(-1.5, 1.5, length=width)
    y = range(-1.5, 1.5, length=height)
    z = [complex(x, y) for y in y, x in x]
    fractal = [julia_set(z[i, j], c, max_iter) for i in 1:height, j in 1:width]
    return fractal
end

# Set parameters
width, height = 1000, 1000
c = -0.4 + 0.6im
max_iter = 100

# Generate the fractal
fractal = create_julia_fractal(width, height, c, max_iter)

# Plot t
```
